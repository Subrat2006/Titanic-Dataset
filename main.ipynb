# Titanic Dataset

## Introduction to the Dataset:

The Titanic Dataset is a Dataset available on [Kaggle](https://www.kaggle.com/azeembootwala/titanic). In this Dataset, the data of how many people died, and how many survived is given with their names.

This Dataset is now-a-days very popular among the Data Scientists.

Now, Let us make the Dataset in this file:

## Importing the modules and the dataset:

We will be importing the Module named "Pandas", and would be working on it by importing the Data, present in the CSV file, which is in the same folder, in which this file is.

Let's first import the module:

```python
import pandas as pd
```

Now, the module is imported.

Now, we will import the dataset in a variable called "df":

```python
df = pd.read_csv("Titanic Dataset.csv")
```

Let's check whether the dataset is imported properly or not:

```python
df
```

So, the dataset is imported successfully in a variable called "df".

So, for calling up the dataset, we will call the variable "df" now.

## Processing and reading the data:

Now that we have imported all the modules and datasets, we'll now process the data:

First of all, let us get the details of the Dataset:

```python
df.describe()
```

```python
df.info()
```

As seen above, we have got some of the major details of the data for the Titanic Dataset.

Now, we'll sort the data according to the age

```python
df_1 = df.sort_values(by = "Age", ascending = True)
```

Now, the data is sorted acccording to the age in a variable "df_1".

Let us call the variable and see if it's working or not.

```python
df_1
```

Now, we have sorted the data according to the age (in ascending order).

So, we don't need the columns Cabin & Embarked. Let's eliminate them:

```python
df_1 = df_1.drop(['Cabin', 'Embarked'], axis = 1)
```

Now, we have eliminated the 2 columns. Let's have a check:

```python
df_1
```

Now, we can see that their are some null values in the age group (as seen in the last part). So, we'll fill the null values by the average values of the total age.

Let's do the same:

```python
df_1['Age'] = df_1['Age'].fillna(df_1['Age'].mean())
```

It's done. 

Let's try it:

```python
df_1['Age']
```

Now, we can see that the null values have been filled with the average values.

Let's see it in the data:

```python
df_1
```

Now, the last part is occupied with the null values filled with average values. So, we need to sort it again.

Let's sort it:

```python
df_2 = df_1.sort_values(by = "Age", ascending = True)
```

It has been sorted and it had been kept in a variable "df_2".

Let's check it:

```python
df_2
```

So, we can see that the data has been sorted in a systematical manner now.

Now, let's see that who were the Top 10 youngest passengers aboard the Titanic:

```python
df_2.head(10)
```

So, above is the list of the Top 10 youngest passengers aboard the Titanic.

Now, let's see the Top 10 eldest passengers aboard the Titanic:

```python
df_2.tail(10)
```

So, we can see above that these were the Top 10 eldest passengers aboard the Titanic.

Now, we don't want the 'Ticket', 'SibSp', and 'Parch' column to be there. So, Let's remove them from our Dataset:

```python
df_2 = df_2.drop(['Ticket', 'SibSp', 'Parch'], axis = 1)
```

Now, Let's check that it has been dropped or not:

```python
df_2
```

So, we have done the task as shown above

Now, Let's sort the data in descending order in terms of Fare:

```python
df_3 = df_2.sort_values(by = "Fare", ascending = False)
```

Now, it's sorted. Let's check the same:

```python
df_3
```

Now, we will get the data of Top 10 highest payers aboard in the Titanic:

```python
df_3.head(10)
```

Now, Let's get the data of the Top 10 lowest payers aboard the Titanic:

```python
df_3.tail(10)
```

Now, we'll get the overall data of the whole processed data:

```python
df_2.describe()
```

```python
df_2.info()
```

Above, we have described the data too.

Now, let's see thatn is there any null value in the dataset:

```python
df_2.isnull()
```

Now, there is no null value.
